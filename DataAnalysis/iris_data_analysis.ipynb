# 1. Loading and Exploring the Iris Dataset

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, 
                          iris.target_names)
```

# 2. Basic Data Analysis

```python
df.describe()
```

```python
df.groupby('species').mean()
```

#### Insights

# 3. Data Visualization

### (i) Line plot - Petal Length by Species

```python
df.groupby('species')['petal length (cm)'].mean().plot(kind='line', marker='o')
```

### (ii) Bar Chart - Sepal Width by Species

```python
df.groupby('species')['sepal width (cm)'].mean().plot(kind='bar')
```

### (iii) Histogram - Distribution of Petal Length

```python
df['petal length (cm)'].plot(kind='hist', bins=20)
```

 ### (iv) Scatter Plot - Sepal vs Petal Length

```python
import seaborn as sns
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
```
